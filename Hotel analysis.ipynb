# Business Problem
In recent years, City Hotel and Resort Hotel have seen high cancellation rates. Each hotel is now dealing with a number of issues
as a rsult, including fewer revenues and less than ideal hotel room use. Consequently, lowering cancellation rates is both hotels. Primary
goal in order to increase their efficiency in gathering revenue, and for us to offer through business advice to adress this problem.

# Research Question
* What are the variables that affect hotel reservation cancellations?
* How can we make hotel reservations cancellations better?
* How will hotels be assisted in making pricing and promotional decisions?

# Data Analysis & Data Visualizations

```python
cancelled_perc=df["is_canceled"].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5,5)) #plotting the percentage of cancelled and not cancelled bookings
plt.bar(["Not Cancelled","Cancelled"],df["is_canceled"].value_counts(),color=["red","black"])
plt.title("Cancelled vs Not Cancelled")
plt.show()
```

```python
from matplotlib import legend


plt.Figure(figsize=(8,5))
ax1=sns.countplot(x=df["hotel"],hue=df["is_canceled"],palette="Blues")
legend=ax1.legend(["Not Cancelled","Cancelled"],title="Status")
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Hotel wise cancellations")
plt.show()
```

```python
resort_hotel=df[df["hotel"]=="Resort Hotel"]
city_hotel=df[df["hotel"]=="City Hotel"]
```

```python
resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
city_hotel["is_canceled"].value_counts(normalize=True)
```

```python
resort_hotel_g=resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel_g=city_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
plt.figure(figsize=(15,5))
plt.title("avg daily rate of resort hotel and city hotel",fontsize=15)
plt.plot(resort_hotel_g.index,resort_hotel_g["adr"],label="Resort Hotel")
plt.plot(city_hotel_g.index,city_hotel_g["adr"],label="City Hotel")
plt.xlabel("Date",fontsize=15)
plt.ylabel("Average Daily Rate",fontsize=15)
plt.legend()
plt.show()
```

```python
df["month"]=df["reservation_status_date"].dt.month
plt.Figure(figsize=(15,8))
ax1=sns.countplot(x="month",hue=df["is_canceled"],data=df,palette="Blues")
plt.legend(["Not Cancelled","Cancelled"],title="Status")
plt.title("Monthly cancellations")
plt.xlabel("Month")
plt.ylabel("Count")
plt.show()
```

```python
plt.Figure(figsize=(15,8))
plt.title("ADR per month",fontsize=15)
sns.barplot(x="month",y="adr",data=df[df["is_canceled"]==1].groupby("month")[["adr"]].sum().reset_index(),palette="Blues")
plt.show()
```

```python
cancelled_data=df[df["is_canceled"]==1]
```

```python
top_10_hotels=cancelled_data["country"].value_counts()[:10]
```

```python
plt.figure(figsize=(10,8))
plt.pie(top_10_hotels,labels=top_10_hotels.index,autopct="%1.1f%%")
plt.title("Top 10 countries with most cancellations")
plt.show()
```

```python
df["market_segment"].value_counts(normalize=True)
```

```python
cancelled_data["market_segment"].value_counts(normalize=True)
```

```python
cancelled_data_adr=cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
cancelled_data_adr.reset_index(inplace=True)
cancelled_data_adr.sort_values("reservation_status_date",inplace=True)
```

```python
not_cancelled_data=df[df["is_canceled"]==0]
not_cancelled_data_adr=not_cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
not_cancelled_data_adr.reset_index(inplace=True)
not_cancelled_data_adr.sort_values("reservation_status_date",inplace=True)
```

```python
plt.figure(figsize=(15,8))
plt.title("Average Daily Rate of cancelled and not cancelled bookings",fontsize=15)
plt.plot(cancelled_data_adr["reservation_status_date"],cancelled_data_adr["adr"],label="Cancelled")
plt.plot(not_cancelled_data_adr["reservation_status_date"],not_cancelled_data_adr["adr"],label="Not Cancelled")
plt.xlabel("Date",fontsize=15)
plt.ylabel("Average Daily Rate",fontsize=15)
plt.legend()
plt.show()
```

```python
cancelled_data_adr=cancelled_data_adr[(cancelled_data_adr["reservation_status_date"]>="2015-07") & (cancelled_data_adr["reservation_status_date"]<="2017-8")]
not_cancelled_data_adr=not_cancelled_data_adr[(not_cancelled_data_adr["reservation_status_date"]>="2015-07") & (not_cancelled_data_adr["reservation_status_date"]<="2017-8")]
```

```python
plt.figure(figsize=(15,8))
plt.title("Average Daily Rate of cancelled and not cancelled bookings",fontsize=15)
plt.plot(cancelled_data_adr["reservation_status_date"],cancelled_data_adr["adr"],label="Cancelled")
plt.plot(not_cancelled_data_adr["reservation_status_date"],not_cancelled_data_adr["adr"],label="Not Cancelled")
plt.xlabel("Date",fontsize=15)
plt.ylabel("Average Daily Rate",fontsize=15)
plt.legend()
plt.show()
```
